```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("uber.csv")

df.head(5)
```

```python
df.tail()
```

```python
df["Date/Time"]= pd.to_datetime(df["Date/Time"])

df["Day"] = df["Date/Time"].apply(lambda x: x.day)
df["Hour"] = df["Date/Time"].apply(lambda x: x.hour)
df["Weekday"] = df["Date/Time"].apply(lambda x: x.weekday())
df.head(5)
```

```python
df.info()
```

```python
df.shape
```

```python
df.tail()
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(df.Day, width =0.6, bins = 30)
plt.title("Density of trips per Day",fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("Density of rides", fontsize = 15)
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(df.Weekday, width = .6, range = (0,6.5),bins = 7, color = "green")
plt.title("Density of trips per Weekday", fontsize = 15)
plt.xlabel("Weekday", fontsize = 15)
plt.ylabel("Density of rides" ,fontsize = 15) 
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(df.Hour, width = .6,bins = 24, color = "orange")
plt.title("Density of trips per Hour", fontsize = 15)
plt.xlabel("Hours", fontsize = 15)
plt.ylabel("Density of rides" ,fontsize = 15) 
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.scatter(df.Lon,df.Lat, color = "purple")
plt.title("Density of trips per Hour", fontsize = 15)
plt.xlabel("Hours", fontsize = 15)
plt.ylabel("Density of rides" ,fontsize = 15) 
```
